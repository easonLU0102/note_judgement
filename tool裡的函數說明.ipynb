```python
#例如在毒品相關判決書中判斷是否為 販賣二級毒品之案件
def re_sell_drug2(txt,name):
    con = [
        '被告[又]{0,1}.{0,30}[共同]{0,2}販賣[第]{0,1}二級毒品',
        '甲○○[又]{0,1}.{0,30}[共同]{0,2}販賣[第]{0,1}二級毒品',
        name+'[又]{0,1}.{0,30}[共同]{0,2}販賣[第]{0,1}二級毒品',
        '基於販賣.{0,20}第二級毒品',
        '被告[又]{0,1}[共同]{0,2}販賣.{1,10}、[第]{0,1}二級毒品',
        '甲○○[又]{0,1}[共同]{0,2}販賣.{1,10}、[第]{0,1}二級毒品',
        name+'[又]{0,1}[共同]{0,2}販賣.{1,10}、[第]{0,1}二級毒品',
            ]
    regex = '|'.join(con)
    print('正則表達式如下\n')
    print(regex,'\n')
    result = re.search(regex,txt)
    return result

txt2 = read_txt(path,2)
result = re_sell_drug2(txt2,name)
print(result)
if result:  #同 result != None:
    print('是販賣二級毒品')
```

# get_digits 篩出文檔內的數值

### 可先找出文檔內目標句子(段落)後再使用
### input string -->output list 

```python
#測試資料
test ='''0.1234公克,1.3公克,234公克,玖仟伍佰貳拾柒公克,玖仟伍佰貳拾柒點五七公克,25公克,
3千元,五十三萬六千元,53萬元,123,000元,五千4百三十元,五千430元,五千400元,五十三萬4千3百元,
十二萬三千元,十六億零六百元,新臺幣壹仟元,
4千元,50萬4千元,玖仟伍佰貳拾柒萬捌仟陸佰捌拾玖元,壹拾柒萬零參佰元
'''
print('不指定單位:',get_digits(test,end='\w'))
print('指定元',get_digits(test,end='元'))
print('指定公克',get_digits(test,end='公克'))
```
